```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

Serie Anual 

```python
# SÃ©rie de 1980 a 2020
np.random.seed(10)
dados1 = np.random.normal(0,1,41)
dados1
```

```python
type(dados1)
```

```python
serie = pd.Series(dados1)
```

```python
type(serie)
```

```python
serie.plot()
plt.show()
```

```python
dados1 = pd.DataFrame(dados1)
dados1
```

```python
dados1.columns = ['valores']
dados1.head()
```

```python
dados1.shape
```

```python
dados1.describe()
```

```python
indice = pd.date_range('1980', periods = len(dados1), freq = 'Y')
indice
```

```python
serie1 = pd.Series(dados1['valores'].values, index = indice)
```

```python
type(serie1)
```

```python
serie1.plot()
plt.show()
```

```python
import scipy.stats as stats
```

```python
stats.probplot(serie1, dist="norm", plot=plt)
plt.title("Normal QQ plot")
plt.show()
```

**Teste Shapiro-Wilk**

CRITÃRIOS:

NÃVEL DE SIGNIFICÃNCIA DE 0,05 ou 5% (MAIS UTILIZADO)

QUANDO p > 0,05 (distribuiÃ§Ã£o normal).

```python
e, p = stats.shapiro(serie1)
print('EstatÃ­stica de teste: {}'.format(e))
print('p-valor: {}'.format(p))
```
